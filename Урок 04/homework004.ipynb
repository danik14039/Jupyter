```python
def new_date(str1):
    return str1[:6]
```

```python

ds['new_date'] = ds['date'].apply(new_date)
ds.head()
```

```python
ds['sqft_price'] = np.round(ds['price']/ds['sqft_lot'],2)
ds.head()
```

```python
data=ds.groupby('new_date')['sqft_price'].agg(mean='mean').reset_index().sort_values(by='new_date')
data
```

```python
plt.figure(figsize=(10,6))

plt.plot(data['new_date'], data['mean'])
plt.title('График средней цены')
plt.xlabel('Дата')
plt.ylabel('Ср. цена за кв.м')
```

График показывает, что средняя стоимость за квадратный метр недвижимости началась снижаться с октября 2014 года по январь 2015 и в январе 2015 года была на минимуме. Затем наблюдается резкий рост до максимума в мае 2015 года.

#### 4.3 Работа с датасетом "house price"
4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
    - Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Для начала, чтобы выявить зависимости воспользуемся матрицей корреляций

Подготовим данные, используем только цифровые значения и округлим до 2 знаков после запятой

```python
corr_matrix = ds.corr(numeric_only=True).round(2)
corr_matrix
```

Выведем матрицу корреляций в виде coolwarm, где значения будут приближены в 1-му значит зависимость есть

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=0.6)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

По матрице корреляций наблюдается, что на цену недвижимости влияют следующие факторы: квадратура жилой площади (sqft_living), уровень конструкций (grade), квадратура жилой площади выше уровня земли (sqf_living_above), количество ванных комнат (bathrooms)

Посмотрим зависимость между ценой и количеством ванных комнат

```python
sns.jointplot(x=ds['price'], y=ds['bathrooms'], kind='reg')
```

По графику видно, что цена увеличивается с увеличением количества ванных комнат

Посмотрим зависимость между ценой и жилой квадратурой

```python
sns.scatterplot(x=ds['price'], y=ds['sqft_living'])
```

Дома со стоимостью от 100 000 до 300 000 Евро имеют квадратуру от 2000 до 6000 кв.метров

```python
sns.barplot(x=ds['grade'], y=ds['price'])
```

График показывает, что чем выше уровень конструкций, тем выше стоимость домов.

```python
sns.jointplot(x=ds['price'], y=ds['sqft_above'], kind='reg')
plt.show()
```

Существует прямая зависимость между ценой и домой с большей квадратурой выше земли. ГРафик показывает, что основное скопление зависимости существует в диапазоне цен до 100 000 Евро и квадратурой от 500 до 4000 кв.метров
